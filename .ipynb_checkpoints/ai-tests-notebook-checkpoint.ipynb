```python
# Run the offline eval (recall@K, MRR) on the smoke dataset

from pathlib import Path

from evals.run_eval import main as eval_main

# Emulate CLI args inside notebook
import sys

argv_prev = sys.argv[:]
try:
    sys.argv = [
        "run_eval",
        "--dataset",
        str(Path("ai-tests/evals/dataset_smoke.json")),
    ]
    rc = eval_main()
finally:
    sys.argv = argv_prev

print("exit_code:", rc)
```

```python
# Quick resume/job similarity sanity check (shared matcher package)

from resume_job_matcher import cosine_sim, embed_texts, load_model_and_tokenizer, sigmoid

resume = "Senior software engineer. Strong Python, FastAPI, PostgreSQL, Docker."
job_good = "Backend engineer — Python, APIs, Postgres, AWS."
job_bad = "Line cook for busy downtown restaurant. Nights and weekends."

model, tokenizer = load_model_and_tokenizer()
embs = embed_texts(model, tokenizer, [resume, job_good, job_bad])

sim_good = cosine_sim(embs[0:1], embs[1:2])
sim_bad = cosine_sim(embs[0:1], embs[2:3])

print(json.dumps(
    {
        "cosine_similarity": {"good": sim_good, "bad": sim_bad},
        "sigmoid_score": {"good": sigmoid(sim_good), "bad": sigmoid(sim_bad)},
    },
    indent=2,
))
```
